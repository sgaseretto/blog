Setting up a blog could be very complicated and downting for people who just want to start publishing articles without having to strugle with setting up a server and all the different solutions available to build one. I'm looking at you Wordpress

Also there are some other options like **Medium**, where you can start writing an article from the get go. I first liked a lot **Medium**, but they started to be very agressive with their member articles, that require you to subscribe to read some of their articles (I know, they have to earn money somehow), but that made me stop considering it as my go to solution to host my own blogs.

Since I'm a big fan of [Jeremy Howards](https://twitter.com/jeremyphoward) courses and tools, I found that he started developing Fastpages with [Hamel Husain](https://twitter.com/hamelhusain?lang=es), and I gave it a try. And so far it is awesome.

Fastpages allows you to write blogposts using Jupyter Notebooks, which is awesome if you are doing a lot of Data Science and Machine Learning stuff. I could write an article of how to setup your own blog with fastpages, but since lot's of people have done this before, I'll share with you a video tutorial about it.
